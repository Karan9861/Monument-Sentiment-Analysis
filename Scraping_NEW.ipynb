```python
import os, time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC 
from selenium.webdriver.common.keys import Keys
import traceback
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd

executable_path = "./chromedriver.exe"
os.environ["webdriver.chrome.driver"] = executable_path

chrome_options = Options()

chrome_prefs = {}
chrome_options.experimental_options["prefs"] = chrome_prefs

chrome_prefs["profile.default_content_settings"] = { "popups": 1 }
chrome_options.add_argument('--disable-blink-features=AutomationControlled')

chrome_options.add_argument('start-maximized')

chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36")
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option("excludeSwitches", ["enable-logging"])

driver = ''

        
def scraper():
    driver = webdriver.Chrome(executable_path=executable_path, options=chrome_options)
    
    driver.get('https://www.google.com/maps/')
    
    
    monument_list = [
        'Nahargarh Fort - Jaipur'
        ]
    
    
    failed_mon = []
    
    for monument in monument_list:
        try:     
            search_bar = WebDriverWait(driver, 40).until(
                        EC.presence_of_element_located((By.XPATH, "//input[@id='searchboxinput']")))

            search_bar.click()
            search_bar.clear()
            search_bar.send_keys(monument)
            search_bar.send_keys(Keys.RETURN)

            time.sleep(3)
        
            rating = WebDriverWait(driver, 30).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[1]/div[1]/div[2]/div/div[1]/div[2]/span[1]/span'))).text
            
            no_of_reviews = WebDriverWait(driver, 30).until( 
            EC.presence_of_element_located((By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[1]/div[1]/div[2]/div/div[1]/div[2]/span[2]/span[1]')))
            
            reviews_count = no_of_reviews.text
            
            no_of_reviews.click()
                    
            sort = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[7]/div[2]/button')))
            time.sleep(3)
            sort.click()
            time.sleep(2)
            
            actions = ActionChains(driver)
            actions.send_keys(Keys.ARROW_DOWN)
            actions.send_keys(Keys.ARROW_UP)
            actions.send_keys(Keys.ENTER)
            actions.perform()
            
            actions = ActionChains(driver)
            actions.send_keys(Keys.TAB * 35)
            actions.perform()
            
            for i in range(110):
                actions = ActionChains(driver) 
                actions.send_keys(Keys.END)
                time.sleep(5)
                actions.perform()
            
            time.sleep(5)
            more = driver.find_elements_by_class_name('w8nwRe')
            for i in more:
                time.sleep(1)
                i.click() 
            
            review = ''
            reviews_count = ''
            time.sleep(5)
            
            
            time.sleep(5)
            user_name = driver.find_elements_by_class_name('d4r55') 
            reviewed_on = driver.find_elements_by_class_name('rsqaWe') 

            review = driver.find_elements_by_class_name('wiI7pd')
            
        
            no_of_stars = driver.find_elements_by_class_name('kvMYJc') 
           
            usr = []
            rev = []
            strs = []
            dur = []
            mon = []
            
            for i, j,k,l in zip(review,no_of_stars,user_name,reviewed_on):
                rev.append(i.text)
                strs.append(j.get_attribute('aria-label'))
                usr.append(k.text)
                dur.append(l.text)
            
            for i in range(len(rev)):
                mon.append(monument)

            reviews = pd.DataFrame({'Monument': mon,'Username': usr, 'Reviewtime':dur,'Review': rev, 'Rating': strs})
            
            print('-'*10)
          
            print(rating)
            reviews.to_csv('jantarmantar.csv',mode='a', index=False, header=False)
            print(reviews_count)

            print('Done Scraping...', monument)
            driver.back()
        
        except:
            
            try:
                more = driver.find_elements_by_class_name('w8nwRe')
                for i in more:
                    time.sleep(1)
                    i.click() 

                review = ''
                reviews_count = ''
                time.sleep(5)


                time.sleep(5)
                user_name = driver.find_elements_by_class_name('d4r55') 
                reviewed_on = driver.find_elements_by_class_name('rsqaWe') 

                review = driver.find_elements_by_class_name('wiI7pd')


                no_of_stars = driver.find_elements_by_class_name('kvMYJc') 

                usr = []
                rev = []
                strs = []
                dur = []
                mon = []

                for i, j,k,l in zip(review,no_of_stars,user_name,reviewed_on):
                    rev.append(i.text)
                    strs.append(j.get_attribute('aria-label'))
                    usr.append(k.text)
                    dur.append(l.text)

                for i in range(len(rev)):
                    mon.append(monument)

                reviews = pd.DataFrame({'Monument': mon,'Username': usr, 'Reviewtime':dur,'Review': rev, 'Rating': strs})

                print('-'*10)

                print(rating)
                reviews.to_csv('trial.csv',mode='a', index=False, header=False)
                print(reviews_count)

                print('Done Scraping...', monument)
                driver.back()
                
            except:
                try:
                    review = ''
                    reviews_count = ''
                    time.sleep(5)


                    time.sleep(5)
                    user_name = driver.find_elements_by_class_name('d4r55') 
                    reviewed_on = driver.find_elements_by_class_name('rsqaWe') 

                    review = driver.find_elements_by_class_name('wiI7pd')


                    no_of_stars = driver.find_elements_by_class_name('kvMYJc') 

                    usr = []
                    rev = []
                    strs = []
                    dur = []
                    mon = []

                    for i, j,k,l in zip(review,no_of_stars,user_name,reviewed_on):
                        rev.append(i.text)
                        strs.append(j.get_attribute('aria-label'))
                        usr.append(k.text)
                        dur.append(l.text)

                    for i in range(len(rev)):
                        mon.append(monument)

                    reviews = pd.DataFrame({'Monument': mon,'Username': usr, 'Reviewtime':dur,'Review': rev, 'Rating': strs})

                    print('-'*10)

                    print(rating)
                    reviews.to_csv('review1.csv',mode='a', index=False, header=False)
                    print(reviews_count)

                    print('Done Scraping...', monument)
                    driver.back()
                    
                except:
                    print(traceback.format_exc())
                    driver.quit()

                    failed_mon.append(monument)
                    driver = webdriver.Chrome(executable_path=executable_path, options=chrome_options)
                    driver.get('https://www.google.com/maps/')

        print('-'*20)
        print(failed_mon)
    
scraper()

try:
    driver.quit()
except:
    pass
```
